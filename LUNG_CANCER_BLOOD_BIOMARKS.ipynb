```python
import re, requests
import pandas as pd
from io import StringIO



html = requests.get(url).text
m = re.search(r"<pre[^>]*>(.*?)</pre>", html, flags=re.S)
if not m:
    raise ValueError("  <pre> .      .")

tsv_text = m.group(1)

#      (TSV)
raw = pd.read_csv(StringIO(tsv_text), sep="\t")
raw.head()
```
